```python
import yfinance as yf
```

```python
#Initialize Ticker class to download price history for a single symbol
sp500 = yf.Ticker("^GSPC")
```

```python
#Query historical prices
sp500 = sp500.history(period="max")
```

```python
#Print data
sp500
```

```python
#See index
sp500.index
```

```python
#Plot the data
sp500.plot.line(y="Close", use_index=True)
```

```python
#Data cleaning: remove unneeded columns
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
#Target set-up (part 1): "Tomorrow, will the price go up or down?", add "Tomorrow" column
sp500["Tomorrow"] = sp500["Close"].shift(-1)
```

```python
#Print data to check shifted values on new "Tomorrow" column
sp500
```

```python
#Target set-up (part 2): "Is Tomorrow's price greater than Today's price?", add "Target" column
#Use astype to convert the boolean to an int
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)
```

```python
#Print data to check values on new column "Target"
sp500
```

```python
#Remove excess historical data
sp500 = sp500.loc["1990-01-01":].copy()
```

```python
#Print data to check if data was removed
sp500
```

```python
#Initial model: RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier

#n_estimators: number of decision trees to train, higher the better
#min_samples_split: avoid overfit
#random_state: seed of the random, to guarantee same random generation when running the model multiple times
model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

#Split data into train and test sets
train = sp500.iloc[:-100]
test = sp500.iloc[-100:]

#List of columns used to predict
predictors = ["Close", "Volume", "Open", "High", "Low"]

#Fit the model
model.fit(train[predictors], train["Target"])
```

```python
#Measure initial model accuracy
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
```

```python
#Convert preds numpy array to pandas series
import pandas as pd

preds = pd.Series(preds, index=test.index)
```

```python
#Calculate precision score
precision_score(test["Target"], preds)
```

```python
#Plot predictions
combined = pd.concat([test["Target"], preds], axis=1)
```

```python
#Check plot
combined.plot()
```

```python
#Wrap everything done into a function to use on backtesting
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
#Backtest function
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
#Run the backtest function
predictions = backtest(sp500, model, predictors)
```

```python
#See predictions count
predictions["Predictions"].value_counts()
```

```python
#See predictions precision score
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
#See % of days that the market went up to compare with our precision score
predictions["Target"].value_counts() / predictions.shape[0]
```

```python
#Add more predictors to our model to try to improve accuracy, using a variety of rolling averages
horizons = [2,5,60,250,1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()
    
    #Ratio between Close Price and last horizon average Close Price
    ratio_column = f"Close_Ratio{horizon}"
    sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"]

    #Trend: number of days in the horizon that the price went up
    trend_column = f"Trend_{horizon}"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]

    new_predictors += [ratio_column, trend_column]
```

```python
#Remove NaN and print data to see new columns
sp500 = sp500.dropna()
sp500
```

```python
#Update model parameters
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
#Tune predict function with probability
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >= .6] = 1
    preds[preds < .6] = 0
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
#Re-run the backtest function with new predictors
predictions = backtest(sp500, model, new_predictors)
```

```python
#See new predictions count
predictions["Predictions"].value_counts()
```

```python
#See new predictions precision score
precision_score(predictions["Target"], predictions["Predictions"])
```
